```python
BEGIN

    INPUT total

    PRINT "Total: ", total

 

    INPUT cash

 

    IF cash >= total THEN

        change = cash - total

        PRINT "Change: ", change

    ELSE

        PRINT "Insufficient Funds"

    ENDIF

 

END
```
